# Funções

No capítulo anterior, usamos várias funções fornecidas pelo Python, como `int` e `float`, e algumas fornecidas pelo módulo `math`, como `sqrt` e `pow`.
Neste capítulo, você aprenderá como criar suas próprias funções e executá-las.
E veremos como uma função pode chamar outra.
Como exemplos, exibiremos letras de músicas do Monty Python.
Esses exemplos bobos demonstram um recurso importante: a capacidade de escrever suas próprias funções é a base da programação.

Este capítulo também apresenta uma nova instrução, o laço `for`, que é usado para repetir uma computação.

## Definindo novas funções

Uma **definição de função** especifica o nome de uma nova função e a sequência de instruções que são executadas quando a função é chamada. Aqui está um exemplo:

````python
def print_lyrics():
    print("I'm a lumberjack, and I'm okay.")
    print("I sleep all night and I work all day.")
````

`def` é uma palavra-chave que indica que esta é uma definição de função.
O nome da função é `print_lyrics`.
Qualquer coisa que seja um nome de variável legal também é um nome de função legal.

Os parênteses vazios após o nome indicam que esta função não recebe argumentos.

A primeira linha da definição da função é chamada de **cabeçalho** -- o resto é chamado de **corpo**.
O cabeçalho deve terminar com dois pontos e o corpo deve ser recuado (indentado). Por convenção, o recuo é sempre de quatro espaços.
O corpo desta função é composto por duas instruções print; em geral, o corpo de uma função pode conter qualquer número de instruções de qualquer tipo.

Definir uma função cria um **objeto function**, que podemos exibir assim:

````python
print_lyrics
````

A saída indica que `print_lyrics` é uma função que não recebe argumentos.
`__main__` é o nome do módulo que contém `print_lyrics`.

Agora que definimos uma função, podemos chamá-la da mesma forma que chamamos funções internas:

````python
print_lyrics()
````

Quando a função é executada, ela executa as instruções no corpo, que exibem as duas primeiras linhas de "The Lumberjack Song".

## Parâmetros

Algumas das funções que vimos recebem argumentos; por exemplo, quando você chama `abs`, você passa um número como argumento.
Algumas funções recebem mais de um argumento; por exemplo, `math.pow` aceita dois, a base e o expoente.

Aqui está uma definição para uma função que aceita um argumento:

````python
def print_twice(string):
    print(string)
    print(string)
````

O nome da variável entre parênteses é um **parâmetro**.
Quando a função é chamada, o valor do argumento é atribuído ao parâmetro.
Por exemplo, podemos chamar `print_twice` assim:

````python
print_twice('Dennis Moore, ')
````

Executar esta função tem o mesmo efeito que atribuir o argumento ao parâmetro e então executar o corpo da função, assim:

````python
string = 'Dennis Moore, '
print(string)
print(string)
````

Você também pode usar uma variável como argumento:

````python
line = 'Dennis Moore, '
print_twice(line)
````

Neste exemplo, o valor de `line` é atribuído ao parâmetro `string`.

## Chamando funções

Uma vez que você tiver definido uma função, você pode usá-la dentro de outra função.
Para demonstrar, escreveremos funções que imprimem a letra de "The Spam Song" (<https://www.songfacts.com/lyrics/monty-python/the-spam-song>).

> Spam, Spam, Spam, Spam,
> Spam, Spam, Spam, Spam,
> Spam, Spam,
> (Lovely Spam, Wonderful Spam!)
> Spam, Spam,

Começaremos com a seguinte função, que recebe dois parâmetros:


````python
def repeat(word, n):
    print(word * n)
````

Podemos usar esta função para exibir a primeira linha da música, assim:

````python
spam = 'Spam, '
repeat(spam, 4)
````

Para exibir as duas primeiras linhas, podemos definir uma nova função que usa `repeat`:

````python
def first_two_lines():
    repeat(spam, 4)
    repeat(spam, 4)
````

E então chamá-la assim:

````python
first_two_lines()
````

Para exibir as últimas três linhas, podemos definir outra função, que também usa `repeat`:

````python
def last_three_lines():
    repeat(spam, 2)
    print('(Lovely Spam, Wonderful Spam!)')
    repeat(spam, 2)
````

````python
last_three_lines()
````

Por fim, podemos reunir tudo com uma função que imprime o verso inteiro:

````python
def print_verse():
    first_two_lines()
    last_three_lines()
````

````python
print_verse()
````

Quando executamos `print_verse`, ela chama `first_two_lines`, que chama `repeat`, que chama `print`.
São muitas funções.

Claro, poderíamos ter feito a mesma coisa com menos funções, mas o ponto deste exemplo é mostrar como as funções podem trabalhar juntas.

## Repetição

Se quisermos exibir mais de um verso, podemos usar uma instrução `for`.
Aqui está um exemplo simples:

````python
for i in range(2):
    print(i)
````

A primeira linha é um cabeçalho que termina com dois pontos.
A segunda linha é o corpo, que precisa ser recuado.

O cabeçalho começa com a palavra-chave `for`, uma nova variável chamada `i` e outra palavra-chave, `in`.
Ela usa a função `range` para criar uma sequência de dois valores, que são `0` e `1`.
Em Python, quando começamos a contar, geralmente começamos em `0`.

Quando a instrução `for` é executada, ela atribui o primeiro valor de `range` a `i` e, em seguida, executa a função `print` no corpo, que exibe `0`.

Quando chega ao final do corpo, ela retorna ao cabeçalho, e é por isso que essa instrução é chamada de **laço de repetição**.
Na segunda vez no laço de repetição, ela atribui o próximo valor de `range` a `i` e o exibe.
Então, como esse é o último valor de `range`, o laço de repetição termina.

Veja como podemos usar um loop `for` para imprimir dois versos da música:

````python
for i in range(2):
    print("Verse", i)
    print_verse()
    print()
````

Você pode colocar um laço `for` dentro de uma função.
Por exemplo, `print_n_verses` tem um parâmetro chamado `n`, que tem que ser um inteiro, e exibe o número dado de versos:

````python
def print_n_verses(n):
    for i in range(n):
        print_verse()
        print()
````

Neste exemplo, não usamos `i` no corpo do loop, mas tem que haver um nome de variável no cabeçalho de qualquer maneira.

## Variáveis ​​e parâmetros são locais

Quando você cria uma variável dentro de uma função, ela é **local**, o que
significa que ela só existe dentro da função.
Por exemplo, a função a seguir pega dois argumentos, os concatena e exibe o resultado duas vezes:

````python
def cat_twice(part1, part2):
    cat = part1 + part2
    print_twice(cat)
````

Aqui está um exemplo que o utiliza:

````python
line1 = 'Always look on the '
line2 = 'bright side of life.'
cat_twice(line1, line2)
````

Quando `cat_twice` é executada, ela cria uma variável local chamada `cat`, que é destruída quando a função termina.
Se tentarmos exibí-la, obtemos um `NameError`:

````python
%%expect NameError

print(cat)
````

Fora da função, `cat` não é definida.

Parâmetros também são locais.
Por exemplo, fora de `cat_twice`, não existe algo como `part1` ou `part2`.

## Diagramas de pilha

Para manter o controle de quais variáveis ​​podem ser usadas onde, às vezes é útil desenhar um **diagrama de pilha**.
Assim como os diagramas de estado, os diagramas de pilha mostram o valor de cada variável, mas também mostram a função à qual cada variável pertence.

Cada função é representada por um **quadro**.
Um quadro é uma caixa com o nome de uma função do lado de fora e os parâmetros e variáveis ​​locais da função do lado de dentro.

Aqui está o diagrama de pilha do exemplo anterior:

````python
from diagram import make_frame, Stack

d1 = dict(line1=line1, line2=line2)
frame1 = make_frame(d1, name='__main__', dy=-0.3, loc='left')

d2 = dict(part1=line1, part2=line2, cat=line1+line2)
frame2 = make_frame(d2, name='cat_twice', dy=-0.3,
                    offsetx=0.03, loc='left')

d3 = dict(string=line1+line2)
frame3 = make_frame(d3, name='print_twice',
                    offsetx=0.04, offsety=-0.3, loc='left')

d4 = {"?": line1+line2}
frame4 = make_frame(d4, name='print',
                    offsetx=-0.22, offsety=0, loc='left')

stack = Stack([frame1, frame2, frame3, frame4], dy=-0.8)
````

````python
from diagram import diagram, adjust


width, height, x, y = [3.77, 2.9, 1.1, 2.65]
ax = diagram(width, height)
bbox = stack.draw(ax, x, y)
# adjust(x, y, bbox)

import matplotlib.pyplot as plt
plt.savefig('chap03_stack_diagram.png', dpi=300)
````

Os quadros são organizados em uma pilha que indica qual função chamou
qual, e assim por diante. Lendo de baixo, `print` foi chamada por `print_twice`, que foi chamada por `cat_twice`, que foi chamada por `__main__` -- que é um nome especial para o quadro mais alto.
Quando você cria uma variável fora de qualquer função, ela pertence a `__main__`.

No quadro para `print`, o ponto de interrogação indica que não sabemos o nome do parâmetro.

## Rastreamentos

Quando um erro de tempo de execução ocorre em uma função, o Python exibe o nome da função que estava em execução, o nome da função que a chamou e assim por diante, na pilha.
Para ver um exemplo, definirei uma versão de `print_twice` que contém um erro -- ela tenta imprimir `cat`, que é uma variável local em outra função:

````python
def print_twice(string):
    print(cat)            # NameError
    print(cat)
````

Agora, aqui está o que acontece quando executamos `cat_twice`:

````python
# Esta célula informa ao Jupyter para fornecer informações detalhadas de
# depuração quando ocorre um erro de tempo de execução, incluindo o
# rastreamento.

%xmode Verbose
````

````python
%%expect NameError

cat_twice(line1, line2)
````

A mensagem de erro inclui um **rastreamento**, que mostra a função que estava em execução quando o erro ocorreu, a função que a chamou e assim por diante.
Neste exemplo, ela mostra que `cat_twice` chamou `print_twice` e o erro ocorreu em um `print_twice`.

A ordem das funções no rastreamento é a mesma que a ordem dos quadros no diagrama de pilha.
A função que estava em execução está na parte inferior.

## Por que funções?

Pode não estar claro ainda por que vale a pena dividir um programa em
funções.
Há vários motivos:

- Criar uma nova função lhe dá a oportunidade de nomear um grupo de
instruções, o que torna seu programa mais fácil de ler e depurar.

- Funções podem tornar um programa menor eliminando código repetitivo.
Mais tarde, se você fizer uma alteração, você só precisa fazê-la em um lugar.

- Dividir um programa longo em funções permite que você depure as partes
uma de cada vez e então as monte em um todo funcional.

- Funções bem projetadas são frequentemente úteis para muitos programas. Depois de escrever e depurar uma, você pode reutilizá-la.

## Depuração

A depuração pode ser frustrante, mas também é desafiadora, interessante e, às vezes, até divertida.
E é uma das habilidades mais importantes que você pode aprender.

De certa forma, a depuração é como um trabalho de detetive.
Você recebe pistas e precisa inferir os eventos que levaram aos
resultados que vê.

A depuração também é como uma ciência experimental.
Depois de ter uma ideia sobre o que está errado, você modifica seu programa e tenta novamente.
Se sua hipótese estiver correta, você pode prever o resultado da modificação e se aproximar mais de um programa funcional.
Se sua hipótese estiver errada, você precisa criar uma nova.

Para algumas pessoas, programação e depuração são a mesma coisa; ou seja, programação é o processo de depurar gradualmente um programa até que ele faça o que você quer.
A ideia é que você comece com um programa funcional e faça pequenas modificações, depurando-as à medida que avança.

Se você perceber que está gastando muito tempo depurando, isso geralmente é um sinal de que você está escrevendo muito código antes de começar os testes.
Se você der passos menores, poderá descobrir que pode avançar mais rápido.

## Glossário

**definição de função** (*function definition*)**:**
Uma instrução que cria uma função.

**cabeçalho** (*header*)**:**
A primeira linha de uma definição de função.

**corpo** (*body*)**:**
A sequência de instruções dentro de uma definição de função.

**objeto *function*** (*function object*)**:**
Um valor criado por uma definição de função.
O nome da função é uma variável que se refere a um objeto de *function*.

**parâmetro** (*parameter*)**:**
Um nome usado dentro de uma função para se referir ao valor passado como um argumento.

**laço de repetição** (*loop*)**:**
Uma instrução que executa uma ou mais instruções, geralmente repetidas vezes.

**variável local (***local variable:***)**
Uma variável definida dentro de uma função e que só pode ser acessada dentro da função.

**diagrama de pilha** (*stack diagram*)**:**
Uma representação gráfica de uma pilha de funções, suas variáveis ​​e os valores aos quais elas se referem.

**quadro** (*frame*)**:**
Uma caixa em um diagrama de pilha que representa uma chamada de função.
Ela contém as variáveis ​​locais e os parâmetros da função.

**rastreamento** (*traceback*)**:**
Uma lista das funções que estão sendo executadas, exibidas quando ocorre uma exceção.
